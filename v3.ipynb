```python
# =============================================================================
# 5. ПОДГОТОВКА ДАННЫХ ДЛЯ N-BEATSx
# =============================================================================

exogenous_features = ['Hour','Month','Weekday','TypeDay','Temperature',
                      'PredCons','PredGen','Price']
available_exog = [f for f in exogenous_features if f in df_prepared.columns]
print(f"\nИспользуемые экзогенные признаки: {available_exog}")

df_nf = df_prepared[['Date','Volume'] + available_exog].copy()
df_nf = df_nf.rename(columns={'Volume':'y','Date':'ds'})
df_nf['unique_id'] = 1
df_nf['ds'] = pd.to_datetime(df_nf['ds'])
df_nf = df_nf.sort_values('ds').reset_index(drop=True)

print("\nДанные для N-BEATSx:")
print(df_nf.head())
print(f"Диапазон дат: {df_nf['ds'].min()} -> {df_nf['ds'].max()}")
```

```python
# =============================================================================
# 6. РАЗДЕЛЕНИЕ НА ВЫБОРКИ
# =============================================================================

train_end = datetime(2022,12,31,23,0,0)
valid_end = datetime(2023,12,31,23,0,0)
test_start = datetime(2025,1,1,0,0,0)
test_end = datetime(2025,5,31,23,0,0)

df_train = df_nf[df_nf['ds'] <= train_end].copy()
df_valid = df_nf[(df_nf['ds'] > train_end) & (df_nf['ds'] <= valid_end)].copy()
df_test  = df_nf[(df_nf['ds'] >= test_start) & (df_nf['ds'] <= test_end)].copy()

print("\n" + "="*50)
print("РАЗДЕЛЕНИЕ ДАННЫХ")
print("="*50)
print(f"Train: {len(df_train)} записей (до {train_end})")
print(f"Valid: {len(df_valid)} записей ({valid_end.year})")
print(f"Test:  {len(df_test)} записей (с {test_start} по {test_end})")

assert df_train['ds'].max() < df_valid['ds'].min()
assert df_valid['ds'].max() < df_test['ds'].min()
```

```python
# =============================================================================
# 7. ОБУЧЕНИЕ МОДЕЛИ N-BEATSx (с отключением прогресс-баров и замером времени)
# =============================================================================

import time  # для замера времени

model = NBEATSx(
    h=horizon,
    input_size=input_size,
    loss=MAE_torch(),
    scaler_type='robust',
    max_steps=1500,
    val_check_steps=50,
    early_stop_patience_steps=5,
    futr_exog_list=available_exog,
    hist_exog_list=[],
    stat_exog_list=[],
    stack_types=['seasonality', 'trend', 'identity'],
    n_blocks=[3, 3, 3],
    mlp_units=[[512, 512]] * 3,
    n_harmonics=2,
    n_polynomials=2,
    alias='NBEATSx',
    # Параметры для Trainer (отключаем визуальный шум)
    enable_progress_bar=False,
    logger=False,
    enable_model_summary=False
)

nf = NeuralForecast(models=[model], freq='H')
print("Начало обучения...")
start_time = time.time()                     # запоминаем время старта
nf.fit(df=df_train, val_size=len(df_valid))
end_time = time.time()                       # время окончания
print(f"Обучение завершено за {end_time - start_time:.2f} секунд.")
model_name = 'NBEATSx'
```

```python
# =============================================================================
# 8. ИТЕРАТИВНОЕ ПРОГНОЗИРОВАНИЕ (единый прогресс-бар)
# =============================================================================

import os
import gc
import pandas as pd
from datetime import datetime
from tqdm import tqdm

# Подавляем логи PyTorch Lightning (убрать встроенные прогресс-бары)
import logging
logging.getLogger("pytorch_lightning").setLevel(logging.WARNING)

PRED_SAVE_FILE = 'results/predictions_full.csv'
run_prediction = False
df_test_with_pred = None

# Если при предыдущем запуске прогнозы уже были сохранены в CSV,
# можно загрузить их, минуя повторный расчёт (экономия времени).
# Проверка наличия сохранённых прогнозов
if os.path.exists(PRED_SAVE_FILE):
    print(f"Найден файл с сохранёнными прогнозами: {PRED_SAVE_FILE}")
    ans = input("Загрузить сохранённые результаты? (да/нет): ").strip().lower()
    if ans in ['да', 'yes', 'y', 'д']:
        print("Загрузка прогнозов...")
        predictions_df = pd.read_csv(PRED_SAVE_FILE, parse_dates=['ds'])
        df_test_with_pred = df_test.merge(predictions_df, on='ds', how='inner')
        print(f"Загружено {len(predictions_df)} прогнозов, покрыто {len(df_test_with_pred)} часов теста.")
    else:
        print("Будет выполнен новый прогноз.")
        run_prediction = True
else:
    print("Сохранённые прогнозы не найдены. Выполняется новый прогноз.")
    run_prediction = True

if run_prediction:
    # Определим общее количество часов в тестовом периоде
    total_hours = len(df_test)
    # Прогресс-бар на всё время
    pbar = tqdm(total=total_hours, desc="Общий прогресс", unit="ч", bar_format="{l_bar}{bar}| {n_fmt}/{total_fmt} [{elapsed}<{remaining}, {rate_fmt}]")

    current_history = df_train.copy()
    all_predictions = []

    # Прогноз выполняется поблочно (по 24 часа) для ускорения.
    # История обновляется только после полного блока, что даёт компромисс между точностью и скоростью.
    # Разбивка по месяцам не обязательна, но оставим для структуры (можно и убрать)
    for month in range(1, 6):
        month_df = df_test[df_test['ds'].dt.month == month].copy()
        if month_df.empty:
            continue

        month_name = month_df['ds'].iloc[0].strftime('%B')
        # Прогноз для месяца блоками по 24 часа
        test_sorted = month_df.sort_values('ds').reset_index(drop=True)
        step = 24

        for i in range(0, len(test_sorted), step):
            block_end = min(i + step, len(test_sorted))
            block_times = test_sorted.loc[i:block_end-1, 'ds']

            # Будущие экзогенные переменные
            last_hist_time = current_history['ds'].max()
            future_times = pd.date_range(
                start=last_hist_time + pd.Timedelta(hours=1),
                periods=step,
                freq='H'
            )
            futr_block = df_nf[df_nf['ds'].isin(future_times)][['ds'] + available_exog].copy()
            if len(futr_block) < step:
                print(f"⚠️ Не хватает данных для блока с {block_times.iloc[0]}")
                pbar.close()
                break
            futr_block['unique_id'] = 1

            # Ограничиваем глубину истории 2000 записями (~3 месяца),
            # чтобы снизить потребление памяти и ускорить расчёт.
            # Для учёта долгосрочных сезонностей (например, прошлогодних праздников) этого недостаточно,
            # но в данной версии выбран баланс скорости.
            recent_history = current_history.iloc[-2000:]

            # Прогноз
            forecast = nf.predict(
                df=recent_history[['ds', 'y', 'unique_id'] + available_exog],
                futr_df=futr_block
            )

            # Извлечение прогнозов (ожидаем колонку model_name с вектором)
            if model_name in forecast.columns:
                pred_series = forecast[model_name]
                if len(pred_series) >= len(block_times):
                    for j, t in enumerate(block_times):
                        all_predictions.append({'ds': t, 'y_pred': pred_series.iloc[j]})
                else:
                    print(f"Длина прогноза ({len(pred_series)}) меньше числа часов в блоке ({len(block_times)})")
                    pbar.close()
                    break
            else:
                print(f"Не найдена колонка {model_name}. Доступны: {forecast.columns.tolist()}")
                pbar.close()
                break

            # Добавляем фактические значения блока в историю
            actual_block = test_sorted.loc[i:block_end-1, ['ds', 'y', 'unique_id'] + available_exog]
            current_history = pd.concat([current_history, actual_block], ignore_index=True).sort_values('ds')

            # Обновляем прогресс-бар на количество обработанных часов в этом блоке
            pbar.update(len(block_times))
            gc.collect()

        # Небольшой отчёт после месяца
        print(f"✅ {month_name} завершён")

    pbar.close()

    # Объединяем все предсказания
    predictions_df = pd.DataFrame(all_predictions).sort_values('ds')
    df_test_with_pred = df_test.merge(predictions_df, on='ds', how='inner')

    os.makedirs('results', exist_ok=True)
    predictions_df.to_csv(PRED_SAVE_FILE, index=False)
    print(f"\n💾 Прогнозы сохранены в файл {PRED_SAVE_FILE}")

print(f"\n🎯 Итоговое покрытие: {len(df_test_with_pred)} из {len(df_test)} часов теста.")
```

```python
# =============================================================================
# 9. ОЦЕНКА МОДЕЛИ (MAE, RMSE, MAPE, MASE)
# =============================================================================

print("\n" + "="*50)
print("ОЦЕНКА ТОЧНОСТИ ПРОГНОЗА")
print("="*50)

y_true = df_test_with_pred['y'].values
y_pred = df_test_with_pred['y_pred'].values

mae = MAE(y_true, y_pred)
rmse = np.sqrt(mean_squared_error(y_true, y_pred))
mape = MAPE(y_true, y_pred) * 100

# Расчёт MASE
# MASE сравнивает MAE модели с MAE наивного сезонного прогноза (лаг 168 часов).
# Значение MASE < 1 означает, что модель работает лучше наивного подхода.
# Наивный сезонный прогноз: лаг 168 часов (недельная сезонность)
df_test_with_pred['naive'] = df_test_with_pred['y'].shift(168)
naive_errors = abs(df_test_with_pred['y'] - df_test_with_pred['naive']).dropna()
mae_naive = naive_errors.mean()
mase = mae / mae_naive

print(f"MAE  = {mae:.3f} МВт·ч")
print(f"RMSE = {rmse:.3f} МВт·ч")
print(f"MAPE = {mape:.3f} %")
print(f"MASE = {mase:.3f}")

if mape < 1.3:
    print("✅ Требование MAPE < 1.3% выполнено!")
else:
    print("⚠️ Требование MAPE < 1.3% не выполнено, требуется донастройка.")

# Сохраняем метрики
with open('results/metrics.txt', 'w', encoding='utf-8') as f:
    f.write(f"MAE  = {mae:.3f} МВт·ч\n")
    f.write(f"RMSE = {rmse:.3f} МВт·ч\n")
    f.write(f"MAPE = {mape:.3f} %\n")
    f.write(f"MASE = {mase:.3f}\n")
    f.write(f"Дата расчёта: {datetime.now().strftime('%Y-%m-%d %H:%M')}\n")
```
